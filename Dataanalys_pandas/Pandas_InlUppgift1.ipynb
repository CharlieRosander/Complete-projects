#### "Skriv därefter ut den dyraste bilens företagsnamn och pris."

-Sorterar df på price (desecending) och tar första raden (head(1)) och skriver ut enbart company och price

```python
df.sort_values(by=["price"], ascending=False).head(1)[["company", "price"]]
```

#### "Skriv därefter ut data om alla Toyota som ni har i lager."

-Använder df.loc med ett condition för att ta ut all data om company == toyota

```python
df.loc[df["company"] == "toyota"]
```

#### "Räkna totalt antal bilar i lager per bilmärke och skriv ut detta."

-Här används value_counts() för att räkna ut hur många antal det är av varje value i kolumnen company

```python
df["company"].value_counts()
```

#### "Skriv ut data om varje bilmärkes dyraste bil."

Sorterar värderna i "price" högst till lägst och visar kolumnerna company & price samt tar bort alla duplicates från company så varje märke bara visar 1 värde

```python
df.sort_values(by="price", ascending=False)[["company", "price"]].drop_duplicates("company")
```
